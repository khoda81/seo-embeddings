```python
from sentence_transformers import SentenceTransformer, util

# Load the model
model = SentenceTransformer("heydariAI/persian-embeddings")
```

```python
# Query and candidates
query = ["ایرپاد"]
candidates = [
    "آیفون ۱۶",
    "آیفون ۱۵",
    "iphone 14 pro",
    "هندزفری",
    "گوشی همراه",
    "سامسونگ a54",
    "طلا",
    "نقره",
]

# Compute embeddings
query_emb = model.encode(query, convert_to_tensor=True)
cand_emb = model.encode(candidates, convert_to_tensor=True)

# Compute cosine similarity
scores = util.cos_sim(query_emb, cand_emb)[0]

# Rank candidates by similarity
ranked = sorted(zip(candidates, scores.tolist()), key=lambda x: x[1], reverse=True)

print(f"Ranking for query {query[0]}:\n")
for text, score in ranked:
    print(f"{text:15} -> {score:.4f}")
```
